```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
times = 5
accs_mnist = []
accs_cifar10 = []

for t in range(times):
    # Carregar dados MNIST
    with h5py.File(f'MNIST_FedAvg_concept_drift_mnist_1_0.1_{t}.h5', 'r') as f:
        acc_mnist = f['rs_test_acc'][:]
        accs_mnist.append(acc_mnist)

    # Carregar dados CIFAR-10
    with h5py.File(f'Cifar10_FedAvg_concept_drift_mnist_1_0.1_{t}.h5', 'r') as f:
        acc_cifar10 = f['rs_test_acc'][:]
        accs_cifar10.append(acc_cifar10)

# Calcula média (se houver mais de uma execução)
mean_acc_cifar10 = np.mean(accs_cifar10, axis=0)
mean_acc_mnist = np.mean(accs_mnist, axis=0)
```

```python
indexes = np.arange(0, len(acc_cifar10))
plt.plot(indexes, mean_acc_cifar10)
plt.show()
```

```python
diff_list = []
for i in range(len(mean_acc_mnist)-1):
    diff = abs(min(mean_acc_mnist[i+1] - mean_acc_mnist[i], 0))
    diff_list.append(diff)
```
