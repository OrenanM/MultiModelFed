```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
dirs_mnist_0 = [f'MNIST_FedAvg_concept_drift_mnist_1_0.1_{i}.h5' for i in range(10)]
dirs_cifar_0 = [f'Cifar10_FedAvg_concept_drift_mnist_1_0.1_{i}.h5' for i in range(10)]
```

```python
dirs_mnist_1 = [f'MNIST_FedAvg_concept_drift_mnist_0.1_1_{i}.h5' for i in range(1)]
dirs_cifar_1 = [f'Cifar10_FedAvg_concept_drift_mnist_0.1_1_{i}.h5' for i in range(1)]
```

```python
def calculate_mean_acc(dirs):
    accs = []
    if type(dirs) == str:
        dirs = [dirs]
    for dir in dirs:
        with h5py.File(dir, 'r') as f:
            acc = f['rs_test_acc']

            accs.append(acc[:])
    return np.mean(accs, axis=0)
```

```python
accs_mnist0 = calculate_mean_acc(dirs_mnist_0)
accs_mnist1 = calculate_mean_acc(dirs_mnist_1)
rounds = np.arange(0, len(accs_mnist0))
plt.plot(rounds, accs_mnist0)
plt.plot(rounds, accs_mnist1)
plt.show()
```
